```python
hamamozu_musteri_sayisi = df_tahakkuk['sozlesme_hesap_no'].nunique()
gumushacikoy_musteri_sayisi = df_tahakkuk_1['sozlesme_hesap_no'].nunique()
goynucek_musteri_sayisi = df_tahakkuk_2['sozlesme_hesap_no'].nunique()

print(f"Hamamözü: {hamamozu_musteri_sayisi}")
print(f"Gümüşhacıköy: {gumushacikoy_musteri_sayisi}")
print(f"Göynücek: {goynucek_musteri_sayisi}")
```

```python
df_tahakkuk_toplam = pd.concat([df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2], ignore_index=True)
df_tahakkuk_toplam.shape
```

```python
# 1. Eksik (Missing) Değerlerin Tespiti
missing_kwh = df_tahakkuk_toplam['kwh'].isna().sum()
print(f"Eksik kwh Değeri Sayısı: {missing_kwh}")

# 2. Negatif Değerlerin Tespiti
negative_kwh = df_tahakkuk_toplam[df_tahakkuk_toplam['kwh'] < 0]
print(f"Negatif kwh Kayıt Sayısı: {len(negative_kwh)}")
print(f"En Düşük Negatif Değer: {df_tahakkuk_toplam['kwh'].min()} kWh")

# 3. Aşırı Uç (Outlier) Değerlerin Tespiti
# İstatistiksel özet ile aykırı değerleri (Max: 153,576) gözlemleyelim
print("\nkwh Sütunu İstatistiksel Özeti:")
print(df_tahakkuk_toplam['kwh'].describe())

# Belirli bir eşiğin (örneğin 10.000 kWh) üzerindeki aşırı değerlerin sayısı
outliers = df_tahakkuk_toplam[df_tahakkuk_toplam['kwh'] > 10000]
print(f"\n10.000 kWh üzerindeki aşırı uç kayıt sayısı: {len(outliers)}")
```

```python
# Hesap sınıfına göre gruplandırma yaparak kwh istatistiklerini hesaplama
tuketim_ozet_tablo = df_tahakkuk_toplam.groupby('hesap_sinifi')['kwh'].agg(['mean', 'median', 'std']).reset_index()

# Sütun isimlerini daha anlaşılır hale getirelim
tuketim_ozet_tablo.columns = ['Hesap Sınıfı', 'Ortalama (kWh)', 'Medyan (kWh)', 'Standart Sapma']

# Sonuçları ortalamaya göre büyükten küçüğe sıralayalım
tuketim_ozet_tablo = tuketim_ozet_tablo.sort_values(by='Ortalama (kWh)', ascending=False)

print("--- Hesap Sınıfına Göre Tüketim İstatistikleri ---")
display(tuketim_ozet_tablo)
```
